```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.font_manager as fm
```

```python
# ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï
all_fonts = fm.findSystemFonts(fontpaths=None, fontext="ttf")

# "Nanum"ÏúºÎ°ú ÏãúÏûëÌïòÎäî Ìè∞Ìä∏ Ï∞æÍ∏∞
nanum_fonts = [font for font in all_fonts if "Nanum" in font]

# Ìè∞Ìä∏ ÏÑ§Ï†ï
if len(nanum_fonts) > 0:
    font_name = fm.FontProperties(fname=nanum_fonts[0]).get_name()
    plt.rcParams["font.family"] = font_name
else:
    print("NanumÏúºÎ°ú ÏãúÏûëÌïòÎäî Ìè∞Ìä∏Í∞Ä ÏÑ§ÏπòÎêòÏßÄ ÏïäÏïòÏäµÎãàÎã§.")
```

# ÌÇ§ÏõåÎìúÎ°ú ÏõêÌïòÎäî Ï±Ö Ï∞æÍ∏∞!!üöÄ

```python

# ÏÇ¨Ïö©ÏûêÎ°úÎ∂ÄÌÑ∞ ÌÇ§ÏõåÎìú ÏûÖÎ†• Î∞õÍ∏∞
def input_keywords():
    keywords_input = input("ÏõêÌïòÎäî Î¶¨Î∑∞ ÌÇ§ÏõåÎìúÎ•º ÏûÖÎ†•ÌïòÍ≥† ÏóîÌÑ∞Î•º ÎàåÎü¨Ï£ºÏÑ∏Ïöî(Ïó¨Îü¨Í∞ú ÏûÖÎ†• Ïãú ',' ÏúºÎ°ú Íµ¨Î∂ÑÌï¥Ï£ºÏÑ∏Ïöî): ")
    keywords = [keyword.strip() for keyword in keywords_input.split(',')]
    return keywords

# Î¶¨Î∑∞ Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï§ÄÎπÑ
def load_and_prepare_data(file_path):
    data = pd.read_csv(file_path)
    data['comment'] = data['comment'].fillna("")  # Îπà Î¶¨Î∑∞Î•º Îπà Î¨∏ÏûêÏó¥Î°ú ÎåÄÏ≤¥
    return data

# ÌÇ§ÏõåÎìúÏôÄ Î¶¨Î∑∞ Í∞ÑÏùò Ïú†ÏÇ¨ÎèÑ Í≥ÑÏÇ∞
def calculate_similarity(data, keywords):
    keywords_string = " ".join(keywords)  # ÌÇ§ÏõåÎìúÎ•º ÌïòÎÇòÏùò Î¨∏ÏûêÏó¥Î°ú Í≤∞Ìï©
    data.loc[len(data)] = pd.Series({'comment': keywords_string})  # ÌÇ§ÏõåÎìú Ìñâ Ï∂îÍ∞Ä
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(data['comment'])
    cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()
    data['similarity_score'] = np.append(cosine_similarities, 0)  # Ïú†ÏÇ¨ÎèÑ Ï†êÏàò Ï∂îÍ∞Ä
    return data

# Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ ÏÉÅÏúÑ 3Í∂åÏùò Ï±Ö ÏãúÍ∞ÅÌôî
def visualize_top_books(data):
    similarity_scores_sum = data.iloc[:-1].groupby('title')['similarity_score'].sum()
    sorted_books_by_similarity = similarity_scores_sum.sort_values(ascending=False).head(3)
    plt.figure(figsize=(5, 3))
    sns.barplot(x=sorted_books_by_similarity.values, y=sorted_books_by_similarity.index)
    plt.title('Ïù¥ Ï±ÖÎì§ÏùÄ ÏûÖÎ†•Ìïú ÌÇ§ÏõåÎìúÏôÄ Ïú†ÏÇ¨Ìïú Î¶¨Î∑∞Í∞Ä ÎßéÏïÑÏöî')
    plt.xlabel('Total Similarity Score')
    plt.ylabel('Book Title')
    plt.show()
    return sorted_books_by_similarity.index[:3]

# Í∞Å Ï±ÖÎ≥Ñ Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ ÏÉÅÏúÑ 5Í∞ú Î¶¨Î∑∞ Ï∂îÏ∂ú Î∞è Ï∂úÎ†•
def extract_top_reviews(data, top_books):
    print("\nÏûÖÎ†•Ìïú ÌÇ§ÏõåÎìúÏôÄ Ïú†ÏÇ¨Ìïú Î¶¨Î∑∞:\n" + "-" * 200)
    for book in top_books:
        print(f"Top reviews for '{book}':\n" + "-" * 200)
        book_reviews = data[data['title'] == book].sort_values(by='similarity_score', ascending=False).head(5)
        for idx, row in enumerate(book_reviews.itertuples(), 1):
            print(f"{idx}: {row.comment[:200]}... [Score: {row.similarity_score:.2f}]")
        print("\n")

# Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Î¶¨Î∑∞Ïùò ÎπÑÏú® Í≥ÑÏÇ∞ Î∞è ÏãúÍ∞ÅÌôî
def calculate_high_similarity_ratio(data):
    threshold = np.percentile(data['similarity_score'], 75)  # ÏÉÅÏúÑ 25%Î•º Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Î¶¨Î∑∞Î°ú Ï†ïÏùò
    high_similarity_reviews = data[data['similarity_score'] > threshold]
    high_similarity_counts = high_similarity_reviews.groupby('title').size()
    total_review_counts = data.iloc[:-1].groupby('title').size()
    total_review_counts = total_review_counts[total_review_counts >= 20]  # Î¶¨Î∑∞ Í∞ØÏàòÍ∞Ä 20Í∞ú Ïù¥ÏÉÅÏù∏ Í≤ΩÏö∞Îßå Í≥†Î†§
    similarity_ratio = (high_similarity_counts / total_review_counts).dropna().sort_values(ascending=False).head(3)
    plt.figure(figsize=(5, 3))
    sns.barplot(x=similarity_ratio.values, y=similarity_ratio.index)
    plt.title('Ïù¥ Ï±ÖÎì§ÏùÄ ÏûÖÎ†•Ìïú ÌÇ§ÏõåÎìúÏôÄ Ïú†ÏÇ¨Ìïú Î¶¨Î∑∞Ïùò ÎπÑÏú®Ïù¥ ÎÜíÏïÑÏöî')
    plt.xlabel('Ratio of High Similarity Reviews')
    plt.ylabel('Book Title')
    plt.show()

def extract_top_high_similarity_reviews(data):
    # Ïú†ÏÇ¨ÎèÑ ÏÉÅÏúÑ 25%Î•º Í∏∞Ï§ÄÏúºÎ°ú ÏÑ§Ï†ï
    threshold = np.percentile(data['similarity_score'], 75)
    
    # Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Î¶¨Î∑∞Îßå ÏÑ†ÌÉù
    high_similarity_reviews = data[data['similarity_score'] > threshold]

    # Î¶¨Î∑∞ ÏàòÍ∞Ä 20Í∞ú Ïù¥ÏÉÅÏù∏ Ï±ÖÎì§Ïóê ÎåÄÌï¥ÏÑúÎßå Í≥†Î†§
    total_review_counts = data.groupby('title').size()
    books_with_enough_reviews = total_review_counts[total_review_counts >= 20].index

    # Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Î¶¨Î∑∞ ÎπÑÏú® Í≥ÑÏÇ∞
    high_similarity_counts = high_similarity_reviews.groupby('title').size()
    similarity_ratio = (high_similarity_counts / total_review_counts).dropna()
    similarity_ratio = similarity_ratio[similarity_ratio.index.isin(books_with_enough_reviews)]
    
    # Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Î¶¨Î∑∞Ïùò ÎπÑÏú®Ïù¥ ÎÜíÏùÄ ÏÉÅÏúÑ 3Í∞ú Ï±Ö ÏÑ†Ï†ï
    top_3_books_by_ratio = similarity_ratio.sort_values(ascending=False).head(3).index

    # Í∞Å Ï±ÖÎ≥ÑÎ°ú Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ ÏÉÅÏúÑ 5Í∞ú Î¶¨Î∑∞ Ï∂îÏ∂ú Î∞è Ï∂úÎ†•
    print("ÏûÖÎ†•Ìïú ÌÇ§ÏõåÎìúÏôÄ Ïú†ÏÇ¨Ìïú Î¶¨Î∑∞:\n" + "-" * 200)
    for book in top_3_books_by_ratio:
        print(f"\nTop 5 Similar Reviews for '{book}':\n" + "-" * 50)
        book_reviews = high_similarity_reviews[high_similarity_reviews['title'] == book]
        top_reviews = book_reviews.sort_values(by='similarity_score', ascending=False).head(5)
        for idx, row in top_reviews.iterrows():
            print(f"{idx+1}: {row['comment'][:200]}... [Score: {row['similarity_score']:.2f}]")
            

# # Î©îÏù∏ Ìï®Ïàò: ÏúÑÏùò Î™®Îì† Ìï®ÏàòÎ•º ÏàúÏ∞®Ï†ÅÏúºÎ°ú Ìò∏Ï∂ú
# def main():
#     file_path = 'all_reviews_filtered_utf8.csv'  # ÌååÏùº Í≤ΩÎ°ú
#     keywords = input_keywords()  # ÌÇ§ÏõåÎìú ÏûÖÎ†• Î∞õÍ∏∞
#     data = load_and_prepare_data(file_path)  # Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï§ÄÎπÑ
#     data_with_similarity = calculate_similarity(data, keywords)  # Ïú†ÏÇ¨ÎèÑ Í≥ÑÏÇ∞
#     calculate_high_similarity_ratio(data_with_similarity)  # Ïú†ÏÇ¨ÎèÑ ÎÜíÏùÄ Î¶¨Î∑∞ ÎπÑÏú® ÏãúÍ∞ÅÌôî
#     extract_top_high_similarity_reviews(data)  # Ïú†ÏÇ¨ÎèÑ ÎÜíÏùÄ Î¶¨Î∑∞ ÎπÑÏú® top3Ïùò Î¶¨Î∑∞ Î¶¨Ïä§Ìä∏
#     top_books_by_similarity = visualize_top_books(data_with_similarity)  # ÏÉÅÏúÑ 3Í∂å Ï±Ö ÏãúÍ∞ÅÌôî
#     extract_top_reviews(data_with_similarity, top_books_by_similarity)  # ÏÉÅÏúÑ 5Í∞ú Î¶¨Î∑∞ Ï∂úÎ†•


# Î©îÏù∏ Ìï®Ïàò
def main():
    file_path = 'all_reviews_filtered_utf8.csv'  # ÌååÏùº Í≤ΩÎ°ú
    data = load_and_prepare_data(file_path)  # Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï§ÄÎπÑ

    while True:
        keywords = input_keywords()  # ÌÇ§ÏõåÎìú ÏûÖÎ†• Î∞õÍ∏∞
        if 'exit' in keywords:
            print("Î∂ÑÏÑùÏùÑ Ï¢ÖÎ£åÌï©ÎãàÎã§.")
            break
        data_with_similarity = calculate_similarity(data, keywords)  # Ïú†ÏÇ¨ÎèÑ Í≥ÑÏÇ∞
        calculate_high_similarity_ratio(data_with_similarity)  # Ïú†ÏÇ¨ÎèÑ ÎÜíÏùÄ Î¶¨Î∑∞ ÎπÑÏú® ÏãúÍ∞ÅÌôî
        extract_top_high_similarity_reviews(data)  # Ïú†ÏÇ¨ÎèÑ ÎÜíÏùÄ Î¶¨Î∑∞ ÎπÑÏú® top3Ïùò Î¶¨Î∑∞ Î¶¨Ïä§Ìä∏
        top_books_by_similarity = visualize_top_books(data_with_similarity)  # ÏÉÅÏúÑ 3Í∂å Ï±Ö ÏãúÍ∞ÅÌôî
        extract_top_reviews(data_with_similarity, top_books_by_similarity)  # ÏÉÅÏúÑ 5Í∞ú Î¶¨Î∑∞ Ï∂úÎ†•
        

if __name__ == '__main__':
    main()
```

# Î¨∏Ïû•ÏúºÎ°ú ÌÜ†ÌÅ∞Ìôî
Ïù¥Ï†Ñ ÏΩîÎìúÏùò TfidfVectorizerÎäî Î¨∏Ïû•ÏùÑ Îã®Ïñ¥Î°ú Î∂ÑÎ¶¨ÌïòÏßÄÎßå, Ïù¥Î•º ÎπÑÌôúÏÑ±ÌôîÌïòÍ≥† Ï†ÑÏ≤¥ ÏûÖÎ†• Î¨∏ÏûêÏó¥ÏùÑ ÌïòÎÇòÏùò ÌÜ†ÌÅ∞ÏúºÎ°ú Ï≤òÎ¶¨ÌïòÏòÄÎã§. 
Î¨∏Ïû•ÏúºÎ°ú Í≤ÄÏÉâ Ïãú Í∏∞Ï°¥ Î¨∏Îß•ÏùÑ Îçî Ïûò Ïú†ÏßÄÌïòÍ≥†Ïûê Î∞îÍø®ÎäîÎç∞ Ìö®Í≥ºÎäî ÏïÑÏßÅ Ïûò Î™®Î•¥Í≤†Îã§ 

```python
# ÏÇ¨Ïö©ÏûêÎ°úÎ∂ÄÌÑ∞ ÌÇ§ÏõåÎìú ÏûÖÎ†• Î∞õÍ∏∞
def input_keywords():
    keywords_input = input("ÏõêÌïòÎäî Î¶¨Î∑∞ ÌÇ§ÏõåÎìúÎ•º ÏûÖÎ†•ÌïòÍ≥† ÏóîÌÑ∞Î•º ÎàåÎü¨Ï£ºÏÑ∏Ïöî(Ïó¨Îü¨Í∞ú ÏûÖÎ†• Ïãú ',' ÏúºÎ°ú Íµ¨Î∂ÑÌï¥Ï£ºÏÑ∏Ïöî): ")
    keywords = [keyword.strip() for keyword in keywords_input.split(',')]
    return keywords

# Î¶¨Î∑∞ Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï§ÄÎπÑ
def load_and_prepare_data(file_path):
    data = pd.read_csv(file_path)
    data['comment'] = data['comment'].fillna("")  # Îπà Î¶¨Î∑∞Î•º Îπà Î¨∏ÏûêÏó¥Î°ú ÎåÄÏ≤¥
    return data

# ÌÇ§ÏõåÎìúÏôÄ Î¶¨Î∑∞ Í∞ÑÏùò Ïú†ÏÇ¨ÎèÑ Í≥ÑÏÇ∞
def calculate_similarity(data, keywords):
    keywords_string = " ".join(keywords)  # ÌÇ§ÏõåÎìúÎ•º ÌïòÎÇòÏùò Î¨∏ÏûêÏó¥Î°ú Í≤∞Ìï©
    data.loc[len(data)] = pd.Series({'comment': keywords_string})  # ÌÇ§ÏõåÎìú Ìñâ Ï∂îÍ∞Ä
    
    # Ï†ÑÏ≤¥ Î¨∏ÏûêÏó¥ÏùÑ Îã®Ïùº ÌÜ†ÌÅ∞ÏúºÎ°ú Ï≤òÎ¶¨ÌïòÎèÑÎ°ù TfidfVectorizer ÏÑ§Ï†ï
    tfidf_vectorizer = TfidfVectorizer(token_pattern=r'(?u)\b\w+\b')
    
    tfidf_matrix = tfidf_vectorizer.fit_transform(data['comment'])
    cosine_similarities = cosine_similarity(tfidf_matrix[-1], tfidf_matrix[:-1]).flatten()
    data['similarity_score'] = np.append(cosine_similarities, 0)  # Ïú†ÏÇ¨ÎèÑ Ï†êÏàò Ï∂îÍ∞Ä
    return data

# Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ ÏÉÅÏúÑ 3Í∂åÏùò Ï±Ö ÏãúÍ∞ÅÌôî
def visualize_top_books(data):
    similarity_scores_sum = data.iloc[:-1].groupby('title')['similarity_score'].sum()
    sorted_books_by_similarity = similarity_scores_sum.sort_values(ascending=False).head(3)
    plt.figure(figsize=(5, 3))
    sns.barplot(x=sorted_books_by_similarity.values, y=sorted_books_by_similarity.index)
    plt.title('Ïù¥ Ï±ÖÎì§ÏùÄ ÏûÖÎ†•Ìïú ÌÇ§ÏõåÎìúÏôÄ Ïú†ÏÇ¨Ìïú Î¶¨Î∑∞Í∞Ä ÎßéÏïÑÏöî')
    plt.xlabel('Total Similarity Score')
    plt.ylabel('Book Title')
    plt.show()
    return sorted_books_by_similarity.index[:3]

# Í∞Å Ï±ÖÎ≥Ñ Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ ÏÉÅÏúÑ 5Í∞ú Î¶¨Î∑∞ Ï∂îÏ∂ú Î∞è Ï∂úÎ†•
def extract_top_reviews(data, top_books):
    print("\nÏûÖÎ†•Ìïú ÌÇ§ÏõåÎìúÏôÄ Ïú†ÏÇ¨Ìïú Î¶¨Î∑∞:\n" + "-" * 200)
    for book in top_books:
        print(f"Top reviews for '{book}':\n" + "-" * 200)
        book_reviews = data[data['title'] == book].sort_values(by='similarity_score', ascending=False).head(5)
        for idx, row in enumerate(book_reviews.itertuples(), 1):
            print(f"{idx}: {row.comment[:200]}... [Score: {row.similarity_score:.2f}]")
        print("\n")

# Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Î¶¨Î∑∞Ïùò ÎπÑÏú® Í≥ÑÏÇ∞ Î∞è ÏãúÍ∞ÅÌôî
def calculate_high_similarity_ratio(data):
    threshold = np.percentile(data['similarity_score'], 75)  # ÏÉÅÏúÑ 25%Î•º Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Î¶¨Î∑∞Î°ú Ï†ïÏùò
    high_similarity_reviews = data[data['similarity_score'] > threshold]
    high_similarity_counts = high_similarity_reviews.groupby('title').size()
    total_review_counts = data.iloc[:-1].groupby('title').size()
    total_review_counts = total_review_counts[total_review_counts >= 20]  # Î¶¨Î∑∞ Í∞ØÏàòÍ∞Ä 20Í∞ú Ïù¥ÏÉÅÏù∏ Í≤ΩÏö∞Îßå Í≥†Î†§
    similarity_ratio = (high_similarity_counts / total_review_counts).dropna().sort_values(ascending=False).head(3)
    plt.figure(figsize=(5, 3))
    sns.barplot(x=similarity_ratio.values, y=similarity_ratio.index)
    plt.title('Ïù¥ Ï±ÖÎì§ÏùÄ ÏûÖÎ†•Ìïú ÌÇ§ÏõåÎìúÏôÄ Ïú†ÏÇ¨Ìïú Î¶¨Î∑∞Ïùò ÎπÑÏú®Ïù¥ ÎÜíÏïÑÏöî')
    plt.xlabel('Ratio of High Similarity Reviews')
    plt.ylabel('Book Title')
    plt.show()

def extract_top_high_similarity_reviews(data):
    # Ïú†ÏÇ¨ÎèÑ ÏÉÅÏúÑ 25%Î•º Í∏∞Ï§ÄÏúºÎ°ú ÏÑ§Ï†ï
    threshold = np.percentile(data['similarity_score'], 75)
    
    # Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Î¶¨Î∑∞Îßå ÏÑ†ÌÉù
    high_similarity_reviews = data[data['similarity_score'] > threshold]

    # Î¶¨Î∑∞ ÏàòÍ∞Ä 20Í∞ú Ïù¥ÏÉÅÏù∏ Ï±ÖÎì§Ïóê ÎåÄÌï¥ÏÑúÎßå Í≥†Î†§
    total_review_counts = data.groupby('title').size()
    books_with_enough_reviews = total_review_counts[total_review_counts >= 20].index

    # Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Î¶¨Î∑∞ ÎπÑÏú® Í≥ÑÏÇ∞
    high_similarity_counts = high_similarity_reviews.groupby('title').size()
    similarity_ratio = (high_similarity_counts / total_review_counts).dropna()
    similarity_ratio = similarity_ratio[similarity_ratio.index.isin(books_with_enough_reviews)]
    
    # Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ Î¶¨Î∑∞Ïùò ÎπÑÏú®Ïù¥ ÎÜíÏùÄ ÏÉÅÏúÑ 3Í∞ú Ï±Ö ÏÑ†Ï†ï
    top_3_books_by_ratio = similarity_ratio.sort_values(ascending=False).head(3).index

    # Í∞Å Ï±ÖÎ≥ÑÎ°ú Ïú†ÏÇ¨ÎèÑÍ∞Ä ÎÜíÏùÄ ÏÉÅÏúÑ 5Í∞ú Î¶¨Î∑∞ Ï∂îÏ∂ú Î∞è Ï∂úÎ†•
    print("ÏûÖÎ†•Ìïú ÌÇ§ÏõåÎìúÏôÄ Ïú†ÏÇ¨Ìïú Î¶¨Î∑∞:\n" + "-" * 200)
    for book in top_3_books_by_ratio:
        print(f"\nTop 5 Similar Reviews for '{book}':\n" + "-" * 50)
        book_reviews = high_similarity_reviews[high_similarity_reviews['title'] == book]
        top_reviews = book_reviews.sort_values(by='similarity_score', ascending=False).head(5)
        for idx, row in top_reviews.iterrows():
            print(f"{idx+1}: {row['comment'][:200]}... [Score: {row['similarity_score']:.2f}]")
            


# Î©îÏù∏ Ìï®Ïàò
def main():
    file_path = 'all_reviews_filtered_utf8.csv'  # ÌååÏùº Í≤ΩÎ°ú
    data = load_and_prepare_data(file_path)  # Îç∞Ïù¥ÌÑ∞ Î°úÎìú Î∞è Ï§ÄÎπÑ

    while True:
        keywords = input_keywords()  # ÌÇ§ÏõåÎìú ÏûÖÎ†• Î∞õÍ∏∞
        if 'exit' in keywords:
            print("Î∂ÑÏÑùÏùÑ Ï¢ÖÎ£åÌï©ÎãàÎã§.")
            break
        data_with_similarity = calculate_similarity(data, keywords)  # Ïú†ÏÇ¨ÎèÑ Í≥ÑÏÇ∞
        calculate_high_similarity_ratio(data_with_similarity)  # Ïú†ÏÇ¨ÎèÑ ÎÜíÏùÄ Î¶¨Î∑∞ ÎπÑÏú® ÏãúÍ∞ÅÌôî
        extract_top_high_similarity_reviews(data)  # Ïú†ÏÇ¨ÎèÑ ÎÜíÏùÄ Î¶¨Î∑∞ ÎπÑÏú® top3Ïùò Î¶¨Î∑∞ Î¶¨Ïä§Ìä∏
        top_books_by_similarity = visualize_top_books(data_with_similarity)  # ÏÉÅÏúÑ 3Í∂å Ï±Ö ÏãúÍ∞ÅÌôî
        extract_top_reviews(data_with_similarity, top_books_by_similarity)  # ÏÉÅÏúÑ 5Í∞ú Î¶¨Î∑∞ Ï∂úÎ†•
        

if __name__ == '__main__':
    main()
```
