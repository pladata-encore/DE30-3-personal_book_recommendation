```python
import pandas as pd
df1 = pd.read_csv("C:/Users/Playdata2/Desktop/project/data/aladin_raw_data.csv")
df2 = pd.read_csv("C:/Users/Playdata2/Desktop/project/data/kyobo_raw_data.csv")
```

```python
print(df1['isbn'].unique())
print(df2['ISBN'].unique())
print(sum(df1['isbn'].isnull()))
print(sum(df2['ISBN'].isnull()))
```

```python
print(df1.columns)
print(df2.columns)
```

```python
# 같은 title 책
merged_df = pd.merge(df1, df2, on='title')
num_duplicates = merged_df['title'].value_counts().sum()

print("중복된 값 개수:", num_duplicates)
```

```python
# 열 이름 변경
new_column_names = {
    'ISBN': 'isbn',
    'content': 'introduction',
    'autho_info': 'author_info',
    'rev_avg':'total_rating',
    'category':'categories',
    'price':'sale_price',
}

df2_1 = df2.rename(columns=new_column_names)

# 변경된 데이터를 새로운 CSV 파일로 저장
df2_1.to_csv('kyobo_new_col.csv', index=False)
```

```python
# 데이터프레임에 칼럼 추가
df1_null = df1.assign(author_info=df2['autho_info'])
df1_null = df1.assign(book_rev=df2['book_rev'])
df1_null = df1.assign(rev_avg=df2['rev_avg'])
print(df1_null.columns)
```

```python
print(df1_null.columns)
print(sum(df1_null['introduction'].isnull()))
print(sum(df1_null['total_rating'].isnull()))
print(sum(df1_null['rev_avg'].isnull()))
print(sum(df1_null['title'].isnull()))
print(sum(df1_null['sale_price'].isnull()))
print(sum(df1_null['review_count'].isnull()))
print(sum(df1_null['categories'].isnull()))
print(sum(df1_null['isbn'].isnull()))
print(sum(df1_null['page'].isnull()))
# print(sum(df1_null['author_info'].isnull()))
```

```python
# 알라딘 null값에 교보 값 처리
df1_null['introduction'] = df1_null['introduction'].fillna(df2['introduction'])
df1_null['categories'] = df1_null['categories'].fillna(df2['categories'])
df1_null['isbn'] = df1_null['isbn'].fillna(df2['isbn'])
df1_null['page'] = df1_null['page'].fillna(df2['page'])
# df_combined['author_info'] = df_combined['author_info'].fillna(df2['author_info'])
```

```python
print(sum(df1_null['introduction'].isnull()))
print(sum(df1_null['categories'].isnull()))
print(sum(df1_null['isbn'].isnull()))
print(sum(df1_null['page'].isnull()))
print(sum(df1_null['author_info'].isnull()))
```

```python
# 교보 null값에 알라딘 값 처리
df2['introduction'] = df2['introduction'].fillna(df1['introduction'])
df2['categories'] = df2['categories'].fillna(df2['categories'])
df2['isbn'] = df2['isbn'].fillna(df1['isbn'])
df2['page'] = df2['page'].fillna(df1['page'])
df2['author_info'] = df2['author_info'].fillna(df1['author_info'])
```

```python
print(sum(df2['introduction'].isnull()))
print(sum(df2['categories'].isnull()))
print(sum(df2['isbn'].isnull()))
print(sum(df2['page'].isnull()))
print(sum(df2['author_info'].isnull()))
```

```python
concat_df = pd.concat([df1_null, df2], axis=0)
sum(concat_df['isbn'].isnull())
```

```python
# isbn 중 null값인 행 제거
concat_df = concat_df.dropna(subset=['isbn'])
```

```python
concat_df.to_csv('merged_df.csv')
```

```python
import pandas as pd
import re

df = pd.read_csv('merged_df.csv')

# title의 () 제거
def remove_brackets(text):
    return re.sub(r'\([^)]*\)', '', text)
    
df['title'] = df['title'].apply(remove_brackets)
```
