# 객체 합치기(열기준)
aladin_list에 a_isbn 합치기

```python
# 첫 번째 CSV 파일 경로
file1_path = 'A_aladin_list_6_final_utf8.csv'
# 두 번째 CSV 파일 경로
file2_path = 'A_information_etc_final_utf8.csv'
# 첫 번째 CSV 파일 읽기
# 'title'과 함께 필요한 컬럼만 선택하여 읽기
df1 = pd.read_csv(file1_path)
# 두 번째 CSV 파일 읽기
df2 = pd.read_csv(file2_path, usecols=['item_id','a_isbn'])
```

```python
#타입확인
print(df1['item_id'].dtype)
print(df2['item_id'].dtype)
```

```python
#타입변경
#str로 바꾸면 안됨
df1['item_id'] = df1['item_id'].astype(str)
#공백제거
df1['item_id'] = df1['item_id'].str.strip()
df2['item_id'] = df2['item_id'].str.strip()
```

```python
#타입확인
print(df1['item_id'].dtype)
print(df2['item_id'].dtype)
```

```python
merged_df = pd.merge(df1, df2, on='item_id', how='inner')
merged_df
```

```python
cleaned_df = merged_df.drop_duplicates(subset=['item_id'], keep='first', ignore_index=True)
cleaned_df
```

```python
# 'a_isbn'에서 중복된 행만 필터링
duplicated_rows = cleaned_df[cleaned_df.duplicated('a_isbn', keep=False)]

# 결과 출력
duplicated_rows.sort_values(by='a_isbn')

#19세 책 2개 a_isbn수기 추가, 나머지 2개는 단종된 상품이라 isbn을알수없음
```

```python
cleaned_df.to_csv('A_reviews_step1_utf8.csv', index=False, encoding="utf-8")
```

# 리뷰 합치기(열기준)
a_isbn이 추가된 aladin_list에 100자리뷰, my리뷰 추가하기  

```python
# 첫 번째 CSV 파일 경로
file1_path = 'A_reviews_step1_utf8.csv'
# 두 번째 CSV 파일 경로
file2_path = 'A_my_reviews_all_final_utf8.csv'
# 세 번째 CSV 파일 경로
file3_path = 'A_100_reviews_all_final_utf8.csv'
# 첫 번째 CSV 파일 읽기
# 'title'과 함께 필요한 컬럼만 선택하여 읽기
df1 = pd.read_csv(file1_path, usecols=[0,1,2,7])
# 두 번째 CSV 파일 읽기
df2 = pd.read_csv(file2_path)
# 세 번째 CSV 파일 읽기
df3 = pd.read_csv(file3_path)
```

```python
print(len(df1), len(df2), len(df3))
```

```python
#동일 item_id, a_isbn 책삭제
df1 = df1.drop_duplicates(subset=["item_id", "a_isbn"], keep="first", ignore_index=True)

#동일 책, 동일 내용 리뷰삭제
df2 = df2.drop_duplicates(subset=["review_comment", "paper_id"], keep="first", ignore_index=True)
df3 = df3.drop_duplicates(subset=["review_comment", "paper_id"], keep="first", ignore_index=True)
print(len(df1), len(df2), len(df3))
```

```python
#열이름 확인
print(df1.columns)
print(df2.columns)
print(df3.columns)
```

```python
# 타입확인
print(df1['a_isbn'].dtype)
print(df2['a_isbn'].dtype)

print(df1['item_id'].dtype)
print(df3['item_id'].dtype)
```

```python
#df1, df2 a_isbn 기준으로 합치기
#의문1 갯수 차이왜 나는지  
merged_df1_2 = pd.merge(df1, df2, on='a_isbn', how='inner')
print(len(df2), len(merged_df1_2))
```

```python
#df1, df2 a_isbn 기준으로 합치기합치기
merged_df1_3 = pd.merge(df1, df3, on='item_id', how='inner')
print(len(df3), len(merged_df1_3))
```

```python
# merged_df1_2와 df3를 'item_id'를 기준으로 합칩니다.
final_merged_df = pd.concat([merged_df1_2, merged_df1_3], ignore_index=True)
final_merged_df
```

```python
final_merged_df.to_csv('A_reviews_all_final_utf8.csv', index=False, encoding="utf-8")
```
